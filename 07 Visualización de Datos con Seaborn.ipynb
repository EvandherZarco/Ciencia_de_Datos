# ¿Qué es Seaborn?

Seaborn es una biblioteca de visualización de datos en Python construida sobre Matplotlib. Proporciona una interfaz de alto nivel para dibujar gráficos estadísticos atractivos e informativos. Seaborn facilita la creación de gráficos complejos de manera sencilla y rápida, gracias a su integración con Pandas y su enfoque en la visualización estadística.

### Ventajas de Seaborn:
- **Integración con Pandas**: Seaborn se integra perfectamente con Pandas, lo que facilita la visualización de DataFrames.
- **Gráficos Estadísticos**: Proporciona funciones para crear gráficos estadísticos que no están disponibles en Matplotlib.
- **Estilos y Temas**: Ofrece estilos y temas predefinidos para crear gráficos atractivos de manera rápida.
- **Facilidad de Uso**: La sintaxis de Seaborn es intuitiva y fácil de aprender, lo que permite crear gráficos complejos con pocas líneas de código.

Seaborn es una herramienta poderosa para el análisis exploratorio de datos (EDA) y la comunicación de resultados estadísticos a través de visualizaciones claras y efectivas.

# Conjunto de Datos sobre Propinas

El conjunto de datos `tips` es un conjunto de datos de prueba incluido en la biblioteca Seaborn. Este conjunto de datos contiene información sobre las propinas que se dan en un restaurante, y es comúnmente utilizado para demostraciones y ejemplos de análisis de datos y visualización.

### Descripción del Dataset

El dataset `tips` incluye las siguientes columnas:

- **total_bill**: El total de la factura (en dólares) antes de la propina.
- **tip**: La cantidad de la propina (en dólares).
- **sex**: El sexo de la persona que paga la cuenta. Puede ser 'Male' (hombre) o 'Female' (mujer).
- **smoker**: Indica si la persona que paga la cuenta es fumadora ('Yes') o no fumadora ('No').
- **day**: El día de la semana en que se realizó la comida. Los valores pueden ser 'Thur' (jueves), 'Fri' (viernes), 'Sat' (sábado) o 'Sun' (domingo).
- **time**: El momento del día en que se realizó la comida. Puede ser 'Lunch' (almuerzo) o 'Dinner' (cena).
- **size**: El tamaño del grupo que comió en el restaurante.

### Primeras Filas del Conjunto de Datos

A continuación, se muestran las primeras filas del conjunto de datos `tips`:

```python
# Cargamos el conjunto de datos de prueba sobre propinas
df=sns.load_dataset('tips')
df.head()
```

## Gráfico de Distribución

El gráfico de distribución (`displot`) es una herramienta versátil para visualizar la distribución de una o más variables utilizando histogramas, KDE (Kernel Density Estimate) y otros enfoques. Es útil para entender la forma subyacente de la distribución de los datos, identificar posibles outliers o anomalías y observar la densidad de los datos.

```python
# Crear un histograma de la columna 'total_bill'
sns.displot(data=df,x='total_bill')

# Mostrar el gráfico
plt.show()
```

### Configuraciones y Estilos en Seaborn

Seaborn permite personalizar los gráficos de manera similar a Matplotlib, proporcionando una amplia variedad de opciones para ajustar la apariencia y funcionalidad de las visualizaciones. A continuación, se describen algunas configuraciones y estilos que aplicaremos al histograma:

1. **Establecer el Estilo de los Gráficos**: Utilizaremos `sns.set(style="whitegrid")` para establecer un estilo de cuadrícula blanca, que mejora la legibilidad del gráfico.

2. **Configurar el Tamaño del Gráfico**: Utilizaremos `plt.figure(figsize=(10, 6))` para definir el tamaño de la figura, haciendo que el gráfico sea más grande y fácil de interpretar.

3. **Crear el Histograma**: Utilizaremos `sns.histplot` para crear el histograma. Configuraremos los siguientes parámetros:
   - `bins=20`: Define el número de intervalos (bins) en el histograma.
   - `kde=True`: Añade la curva KDE (Kernel Density Estimate) para mostrar la densidad de los datos.
   - `color='skyblue'`: Establece el color del histograma.

4. **Añadir Título y Etiquetas**: Utilizaremos `set_title`, `set_xlabel` y `set_ylabel` para añadir un título al gráfico y etiquetas a los ejes x e y.

5. **Mostrar la Cuadrícula**: Utilizaremos `histograma.grid(True)` para añadir una cuadrícula al fondo del gráfico, mejorando la legibilidad de los datos.

### Ejemplo Completo

A continuación, se muestra el código completo para crear y personalizar el histograma utilizando Seaborn.

```python
# Establecer el estilo de los gráficos
sns.set_theme(style="whitegrid")

# Configurar el tamaño del gráfico
plt.figure(figsize=(10, 6))

# Crear un histograma de la columna 'total_bill' con configuraciones adicionales
histograma = sns.histplot(data=df,
                          x='total_bill',
                          bins=20,
                          kde=True, color='skyblue')

# Configuraciones adicionales
histograma.set_title('Distribución del Total de la Factura', fontsize=16)
histograma.set_xlabel('Total de la Factura ($)', fontsize=14)
histograma.set_ylabel('Frecuencia', fontsize=14)
histograma.grid(True)

# Mostrar el gráfico
plt.show()
```

# Temas Disponibles en Seaborn

Seaborn proporciona varios temas predefinidos que permiten ajustar la apariencia general de los gráficos. Estos temas controlan elementos como los colores, las cuadrículas y los estilos de las líneas, facilitando la creación de visualizaciones atractivas y coherentes. A continuación, se describen los temas disponibles en Seaborn y sus características principales.

## Temas de Seaborn

1. **darkgrid**
   - **Descripción**: Este tema utiliza un fondo oscuro con una cuadrícula gris clara.
   - **Características**: Ideal para gráficos con muchos datos donde la cuadrícula puede ayudar a la legibilidad.
   - **Código de ejemplo**: `sns.set_theme(style="darkgrid")`

2. **whitegrid**
   - **Descripción**: Este tema utiliza un fondo blanco con una cuadrícula gris.
   - **Características**: Útil para gráficos donde la claridad y la simplicidad son importantes.
   - **Código de ejemplo**: `sns.set_theme(style="whitegrid")`

3. **dark**
   - **Descripción**: Este tema utiliza un fondo oscuro sin cuadrícula.
   - **Características**: Atractivo para gráficos minimalistas y presentaciones.
   - **Código de ejemplo**: `sns.set_theme(style="dark")`

4. **white**
   - **Descripción**: Este tema utiliza un fondo blanco sin cuadrícula.
   - **Características**: Proporciona una apariencia limpia y despejada, ideal para gráficos que necesitan destacarse por su simplicidad.
   - **Código de ejemplo**: `sns.set_theme(style="white")`

5. **ticks**
   - **Descripción**: Este tema es similar al tema `white`, pero añade marcas ("ticks") en los ejes.
   - **Características**: Útil para gráficos donde las marcas en los ejes pueden ayudar a la precisión en la lectura de los datos.
   - **Código de ejemplo**: `sns.set_theme(style="ticks")`

# Explorar Diferentes Estilos y Usos de displot

Después de crear un histograma básico utilizando `displot` para visualizar la distribución de la columna `total_bill`, exploraremos ahora cómo aplicar diferentes estilos en Seaborn y utilizar `displot` para mostrar la relación entre dos variables.

Primero, vamos a cambiar el estilo de los gráficos a `dark`, lo que proporciona un fondo oscuro y una apariencia moderna. Luego, utilizaremos `displot` de una manera alternativa para crear un gráfico de distribución conjunta de `total_bill` y `tip`.

### Aplicar el Estilo Dark y Crear un Gráfico de Distribución Conjunto

A continuación, configuramos el estilo a `dark` y creamos un gráfico de distribución conjunta de `total_bill` y `tip`. Esto nos permitirá visualizar cómo se relacionan estas dos variables en el conjunto de datos.

```python
# Establecer el estilo de los gráficos a 'dark'
sns.set_theme(style="dark")

# Crear un gráfico de distribución de 'total_bill' y 'tip'
sns.displot(data=df, x='total_bill', y='tip')

# Mostrar el gráfico
plt.show()
```

# Visualización de Relaciones con relplot

En la sección anterior, utilizamos `displot` para crear un gráfico de distribución conjunta, lo cual es útil para explorar la relación entre dos variables numéricas. Sin embargo, Seaborn ofrece otras funciones para visualizar relaciones entre variables de manera más detallada y flexible.

Una de estas funciones es `relplot`, que es especialmente útil para visualizar la relación entre dos variables. Este tipo de gráfico puede revelar patrones, tendencias y correlaciones que podrían pasar desapercibidos en una representación tabular de los datos.

Para ilustrar esto, primero vamos a cambiar el estilo de los gráficos a `white`, y luego utilizaremos `relplot` para mostrar la relación entre `total_bill` y `tip`.

```python
# Establecer el estilo de los gráficos a 'white'
sns.set_theme(style="white")

# Crear un gráfico de relación entre 'total_bill' y 'tip'
sns.relplot(data=df, x='total_bill', y='tip')

# Mostrar el gráfico
plt.show()
```

# Exploración Avanzada con displot y hue

Después de explorar cómo `relplot` puede ayudarnos a visualizar la relación entre dos variables, regresamos a `displot` para mostrar cómo podemos aprovechar otros parámetros para obtener más información de nuestros datos.

El parámetro `hue` en `displot` permite agregar una dimensión adicional a nuestra visualización mediante la codificación de color. Esto es especialmente útil para ver cómo una variable categórica afecta la distribución de los datos.

### Gráfico de Distribución Conjunto con hue

En el siguiente ejemplo, utilizamos `displot` para crear un gráfico de distribución conjunta de `total_bill` y `tip`, pero esta vez incorporamos el parámetro `hue` para diferenciar entre fumadores y no fumadores. También configuramos el tipo de gráfico a `kde` (Kernel Density Estimate) para visualizar la densidad de las distribuciones.

Además, cambiaremos el estilo del gráfico a `darkgris` para resaltar más los trazos y mejorar la precisión en la lectura de los datos.

```python
# Establecer el estilo de los gráficos a 'ticks'
sns.set_theme(style="darkgrid")

# Crear un gráfico de distribución conjunto con hue
sns.displot(data=df, 
            x='total_bill', 
            y='tip', 
            hue='smoker', 
            kind='kde')

# Mostrar el gráfico
plt.show()
```

# Añadiendo una Barra de Color con displot

Hasta ahora hemos explorado cómo utilizar `displot` para visualizar la relación entre dos variables y cómo incorporar una variable categórica adicional usando `hue`. Ahora, vamos a ver cómo se puede añadir una barra de color a nuestros gráficos con el parámetro `cbar`.

La barra de color (`cbar`) es una herramienta útil para visualizar la densidad de datos en un gráfico, proporcionando una referencia adicional para interpretar la intensidad de las concentraciones de datos.

### Gráfico de Distribución Conjunto con Barra de Color

En este ejemplo, utilizaremos `displot` para crear un gráfico de distribución conjunta de `total_bill` y `tip`, y añadiremos una barra de color para representar la densidad de los datos. La barra de color nos ayudará a entender mejor cómo se distribuyen las concentraciones de puntos en el gráfico.

```python
# Establecer el estilo de los gráficos a 'ticks'
sns.set_theme(style="whitegrid")

# Crear un gráfico de distribución conjunto con barra de color
sns.displot(data=df, x='total_bill', y='tip', cbar=True)

# Mostrar el gráfico
plt.show()
```

# Visualización de Relaciones con Jointplot

Hasta ahora, hemos utilizado `displot` para explorar la distribución conjunta de dos variables y cómo incorporar elementos adicionales como barras de color. Para profundizar en el análisis bivariado, utilizaremos `jointplot`.

## Jointplot: Relaciones Bivariadas y Distribuciones Univariadas

Un `jointplot` es una poderosa herramienta de visualización que muestra simultáneamente la relación entre dos variables y sus distribuciones univariadas a lo largo de los ejes. Este tipo de gráfico es particularmente útil para entender tanto las relaciones individuales como las interacciones entre dos variables.

El `jointplot` crea:
- Un gráfico de dispersión (scatter plot) que muestra la relación entre las dos variables.
- Dos histogramas o gráficos de densidad (uno en cada eje) que muestran la distribución de cada variable individualmente.

Esta combinación proporciona una visión completa de cómo las variables están relacionadas y cómo se distribuyen.

```python
# Crear un jointplot de 'total_bill' y 'tip'
sns.jointplot(data=df, x='total_bill', y='tip')

# Mostrar el gráfico
plt.show()
```

# Añadiendo Dimensiones con hue en Jointplot

Además de mostrar la relación bivariada entre dos variables y sus distribuciones univariadas, `jointplot` también permite incorporar una variable categórica adicional utilizando el parámetro `hue`. Esto nos permite visualizar cómo diferentes categorías afectan la relación entre las dos variables principales.

### Jointplot con hue

A continuación, utilizaremos `jointplot` para mostrar la relación entre `total_bill` y `tip`, pero esta vez diferenciando por el sexo del cliente (`sex`). Esto nos permitirá observar si existen diferencias significativas en las propinas entre hombres y mujeres.

```python
# Crear un jointplot de 'total_bill' y 'tip' con hue por sexo
sns.jointplot(data=df, x='total_bill', y='tip', hue='sex')

# Mostrar el gráfico
plt.show()
```

# Añadiendo Dimensiones y Estilos en Jointplot

Ahora, incorporaremos una variable categórica adicional con `hue` y cambiaremos el tipo de gráfico a KDE para ver las densidades.

```python
# Crear un jointplot de 'total_bill' y 'tip' con hue por sexo y tipo KDE
sns.jointplot(data=df, x='total_bill',
               y='tip', hue='sex', kind='kde')

# Mostrar el gráfico
plt.show()
```

# Visualización de Distribuciones con Boxplot

# Visualización de Tendencias con Gráfico de Líneas

Un gráfico de líneas es una excelente herramienta para visualizar tendencias a lo largo de una variable continua. Podemos hacer uso del parámetro `hue` para diferenciar entre diferentes categorías dentro de los datos, lo que añade una capa adicional de información y hace que la visualización sea más rica.

### Área Sombreada en el Gráfico de Líneas

En el gráfico de líneas generado por Seaborn, el área sombreada alrededor de las líneas representa el intervalo de confianza. Este intervalo proporciona una estimación visual de la variabilidad de los datos y nos da una idea de la precisión de las tendencias mostradas. El intervalo de confianza se calcula automáticamente y suele ser del 95% por defecto. **Este área sombreada indica el rango en el cual esperamos que se encuentren la mayoría de los datos en base a nuestras muestras.**

### Ejemplo de Gráfico de Líneas

En este ejemplo, crearemos un gráfico de líneas que muestra la tendencia del `total_bill` a lo largo de los días de la semana, diferenciando entre fumadores y no fumadores. La inclusión del parámetro `hue` nos permitirá ver cómo estas categorías afectan las tendencias de los datos.

```python
# Crear un gráfico de líneas de 'total_bill' por 'day' con 'hue' por 'smoker'
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='day', y='total_bill', 
             hue='smoker', marker='o')

# Añadir título y etiquetas
plt.title('Tendencia de Total Bill por Día y Estado de Fumador', fontsize=16)
plt.xlabel('Día de la Semana', fontsize=14)
plt.ylabel('Total Bill ($)', fontsize=14)

# Mostrar el gráfico
plt.show()
```

# Visualización de Distribuciones con Swarmplot

Un `swarmplot` es una excelente herramienta para visualizar la distribución de datos categóricos. Cada observación se muestra como un punto, y el gráfico ajusta los puntos para evitar la superposición. Esto permite ver claramente la distribución de los datos y detectar patrones o anomalías.

### Ejemplo de Swarmplot

En este ejemplo, crearemos un `swarmplot` para visualizar la distribución de `total_bill` a través de los días de la semana, diferenciando entre fumadores y no fumadores.

```python
# Crear un swarmplot de 'total_bill' por 'day' con 'hue' por 'smoker'
plt.figure(figsize=(12, 6))
sns.swarmplot(data=df, x='day', 
              y='total_bill', 
              hue='smoker', dodge=False)

# Añadir título y etiquetas
plt.title('Distribución de Total Bill por Día y Estado de Fumador', fontsize=16)
plt.xlabel('Día de la Semana', fontsize=14)
plt.ylabel('Total Bill ($)', fontsize=14)

# Mostrar el gráfico
plt.show()
```

# Visualización con FacetGrid

El `FacetGrid` de Seaborn es una herramienta poderosa para crear múltiples gráficos basados en las combinaciones de variables categóricas. Esto es particularmente útil para comparar visualmente subgrupos dentro del conjunto de datos y entender cómo las relaciones entre las variables pueden variar entre estos subgrupos.

### Ejemplo de FacetGrid

En este ejemplo, crearemos un `FacetGrid` para visualizar la distribución de `total_bill` diferenciando por el día de la semana y el estado de fumador.

```python
# Crear un FacetGrid de 'total_bill' por 'day' y 'smoker'
g = sns.FacetGrid(df, col="day",
                  row="smoker",
                  margin_titles=True)
g.map(sns.histplot, "total_bill", kde=True)

# Añadir título general
g.figure.suptitle('Distribución de Total Bill por Día y Estado de Fumador', fontsize=16)
plt.subplots_adjust(top=0.9)

# Mostrar el gráfico
plt.show()
```

# Visualización Compleja con FacetGrid y Scatterplot

El `FacetGrid` combinado con un `scatterplot` es una herramienta poderosa para explorar relaciones complejas en los datos. Al utilizar los parámetros `hue`, `col` y `row`, podemos visualizar cómo múltiples dimensiones categóricas y continuas interactúan entre sí, proporcionando una visión detallada y rica de los datos. Además, podemos personalizar los colores de los gráficos utilizando el parámetro `palette`, que permite elegir entre varias paletas de colores predefinidas o crear una personalizada.

### Palette en Seaborn

El parámetro `palette` en Seaborn controla los colores utilizados en los gráficos cuando se utilizan categorías. Algunos valores comunes de `palette` incluyen:

- `"deep"`: Paleta de colores profundos, ideal para gráficos con muchos niveles de `hue`.
- `"muted"`: Colores suaves, buenos para presentaciones donde no se necesita un alto contraste.
- `"bright"`: Colores brillantes, útiles para destacar diferencias entre categorías.
- `"pastel"`: Colores pastel, adecuados para gráficos donde se prefiere un aspecto más suave.
- `"dark"`: Colores oscuros, buenos para temas con fondo claro.
- `"colorblind"`: Paleta optimizada para personas con dificultades de visión de colores.

En este ejemplo, utilizaremos la paleta `"Set1"`, que proporciona colores brillantes y distintivos, facilitando la diferenciación entre categorías.

### Ejemplo de FacetGrid con Scatterplot

En este ejemplo, crearemos un `FacetGrid` con `scatterplot` para visualizar la relación entre `total_bill` y `tip`, diferenciando por sexo (`sex`), estado de fumador (`smoker`) y día de la semana (`day`). Utilizaremos la paleta de colores `"Set1"` para mejorar la claridad visual.


```python
# Crear un FacetGrid con scatterplot de 'total_bill' vs 'tip' por 'sex', 'smoker' y 'day'
g = sns.FacetGrid(df, col="day", 
                  row="smoker", hue="sex", 
                  margin_titles=True, palette="Set1")
g.map(sns.scatterplot, "total_bill", "tip", alpha=.7)
g.add_legend()

# Añadir título general
g.fig.suptitle('Relación entre Total Bill y Tip por Día, Estado de Fumador y Sexo', fontsize=16)
plt.subplots_adjust(top=0.9)

# Mostrar el gráfico
plt.show()
```

A continuación, cargaremos el conjunto de datos `penguins` y exploraremos algunas visualizaciones básicas para familiarizarnos con los datos.

```python
# Cargar el conjunto de datos de pingüinos
pinguinos = sns.load_dataset('penguins')

# Mostrar las primeras filas del conjunto de datos
pinguinos.head()
```

# Visualización de Relaciones con Pairplot

El `pairplot` es una herramienta poderosa en Seaborn para visualizar relaciones entre múltiples variables en un conjunto de datos. Este gráfico crea una matriz de gráficos de dispersión para cada par de variables numéricas y gráficos de distribución en la diagonal. Al utilizar el parámetro `hue`, podemos diferenciar entre categorías, añadiendo una dimensión adicional a nuestra exploración visual.

### Ejemplo de Pairplot con Hue

En este ejemplo, utilizaremos `pairplot` para visualizar las relaciones entre las variables numéricas del conjunto de datos `penguins`, diferenciando por especie (`species`). Esto nos permitirá observar cómo las características físicas de los pingüinos varían entre las diferentes especies.

```python
# Crear un pairplot de las variables numéricas, diferenciando por especie
sns.pairplot(data=pinguinos,
             hue='species',
             palette='Set2', markers=['o', 's', 'D'])

# Añadir título general
plt.suptitle('Relaciones entre Variables de Pingüinos por Especie', fontsize=16, y=1.02)

# Mostrar el gráfico
plt.show()
```

# Personalización Avanzada de Pairplot

El `pairplot` es extremadamente versátil y permite una gran cantidad de personalizaciones para adaptar las visualizaciones a nuestras necesidades específicas. Podemos utilizar diferentes tipos de gráficos para las distintas secciones de la matriz de gráficos, como histogramas, gráficos de densidad y gráficos de dispersión. Esta personalización nos ayuda a obtener una visión más completa de los datos y sus relaciones.

### Ejemplo de Pairplot con Personalización

En este ejemplo, crearemos un `pairplot` del conjunto de datos `penguins`, diferenciando por especie (`species`). Personalizaremos los gráficos en las diferentes partes del `pairplot`:
- En la parte superior de la matriz (`upper`), utilizaremos histogramas (`histplot`).
- En la parte inferior de la matriz (`lower`), utilizaremos gráficos de densidad (`kdeplot`).
- En la diagonal (`diag`), utilizaremos histogramas con curvas de densidad superpuestas.

```python
# Crear un pairplot personalizado
g = sns.pairplot(data=pinguinos, hue='species', palette='Set3')

# Mapear diferentes tipos de gráficos en las distintas secciones de la matriz
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot, kde=True)

# Añadir título general
plt.suptitle('Relaciones Personalizadas entre Variables de Pingüinos por Especie', fontsize=16, y=1.02)

# Mostrar el gráfico
plt.show()
```

# Exploración de Distribuciones Categóricas con Catplot

Después de visualizar relaciones complejas entre las variables numéricas en el conjunto de datos de pingüinos, ahora nos centraremos en explorar la distribución de variables categóricas. Para esto, utilizaremos `catplot`, una herramienta versátil en Seaborn que permite crear gráficos categóricos variados como gráficos de barras, gráficos de puntos, gráficos de cajas y más.

### Ejemplo de Catplot

En este ejemplo, utilizaremos `catplot` para visualizar la distribución de la longitud del pico (`bill_length_mm`) entre las diferentes especies de pingüinos. Esto nos ayudará a entender mejor cómo varía esta característica física entre las especies Adelie, Chinstrap y Gentoo.

```python
# Crear un catplot de 'bill_length_mm' por 'species'
sns.catplot(data=pinguinos, x='species', y='bill_length_mm', 
            palette='Set1')

# Añadir título y etiquetas
plt.title('Distribución de la Longitud del Pico por Especie', fontsize=16)
plt.xlabel('Especie', fontsize=14)
plt.ylabel('Longitud del Pico (mm)', fontsize=14)

# Mostrar el gráfico
plt.show()
```

# Visualización de Distribuciones Categóricas con Swarmplot

Hasta ahora, hemos explorado la distribución de la longitud del pico (`bill_length_mm`) entre las diferentes especies de pingüinos utilizando `catplot`. Para una visualización aún más detallada de la distribución de los datos, podemos utilizar el `kind='swarm'` en `catplot`.

Un `swarmplot` muestra cada observación como un punto, ajustando los puntos para evitar la superposición. Esto permite ver claramente la distribución de los datos y detectar patrones o anomalías de una manera más precisa.

```python
# Crear un catplot de 'bill_length_mm' por 'species' con kind='swarm'
sns.catplot(data=pinguinos, x='species', y='bill_length_mm', 
            kind='swarm', palette='Set1')

# Añadir título y etiquetas
plt.title('Distribución de la Longitud del Pico por Especie (Swarmplot)', fontsize=16)
plt.xlabel('Especie', fontsize=14)
plt.ylabel('Longitud del Pico (mm)', fontsize=14)

# Mostrar el gráfico
plt.show()
```

# Exploración de Distribuciones Categóricas con Swarmplot y Hue

Para agregar una dimensión adicional a nuestra visualización, utilizaremos el parámetro `hue` para diferenciar los datos por sexo (`sex`). Esto nos permitirá ver cómo varía la longitud del pico (`bill_length_mm`) entre las diferentes especies de pingüinos y entre machos y hembras.

```python
# Crear un catplot de 'bill_length_mm' por 'species' con kind='swarm' y hue='sex'
sns.catplot(data=pinguinos, x='species', y='bill_length_mm', 
            kind='swarm', hue='sex', palette='Set1')

# Añadir título y etiquetas
plt.title('Distribución de la Longitud del Pico por Especie y Sexo (Swarmplot)', fontsize=16)
plt.xlabel('Especie', fontsize=14)
plt.ylabel('Longitud del Pico (mm)', fontsize=14)

# Mostrar el gráfico
plt.show()
```

# Combinación de Gráficos Violinplot y Swarmplot

Para obtener una visualización más completa de los datos, podemos combinar diferentes tipos de gráficos en una sola visualización. En este ejemplo, utilizaremos un `violinplot` para mostrar la distribución y densidad de la longitud del pico (`bill_length_mm`) entre las diferentes especies de pingüinos, y superpondremos un `swarmplot` para visualizar cada observación individual.

### Detalles de la Combinación

- **Violinplot**: Utilizamos `kind='violin'` en `catplot` para crear un gráfico de violín que muestra la distribución y densidad de los datos. Establecemos `inner=None` para eliminar los gráficos internos del violín, dejando espacio para el `swarmplot`.
- **Swarmplot**: Superponemos un `swarmplot` sobre el `violinplot` para mostrar cada punto de datos individual. Configuramos `color='k'` para que los puntos sean de color negro y `size=3` para ajustar el tamaño de los puntos.

```python
# Crear un catplot de 'bill_length_mm' por 'species' con kind='violin' y inner=None
g = sns.catplot(data=pinguinos, x='species', 
                y='bill_length_mm', kind='violin', 
                inner=None, palette='Set1')

# Superponer un swarmplot al violinplot
sns.swarmplot(data=pinguinos, x='species', 
              y='bill_length_mm', color='k', size=3, ax=g.ax)

# Añadir título y etiquetas
plt.title('Distribución y Densidad de la Longitud del Pico por Especie', fontsize=16)
plt.xlabel('Especie', fontsize=14)
plt.ylabel('Longitud del Pico (mm)', fontsize=14)

# Mostrar el gráfico
plt.show()
```
